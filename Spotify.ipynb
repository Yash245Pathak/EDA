# Converting duration in ms in seconds 

```python
df["duration"]=df["duration_ms"].apply(lambda x:round(x/1000))
df.drop("duration_ms", inplace=True,axis=1)
```

```python
df.duration.head()
```

```python
df.head()
```

# Visualiztion

```python
corr_df=df.drop(["key","mode"],axis=1).corr(method="pearson")
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(corr_df,annot=True,fmt=".1g",center=0,vmin=-1,vmax=1,cmap="inferno",linewidths=1,linecolor="Black")
plt.show()
```

Coorelation b/w loudness and energy on 4% of data

```python
sample_df = df.sample(int(0.004*len(df)))
```

```python
print(len(sample_df))
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "loudness", x="energy", color = "Gray").set(title="Correlation between Loudness and Energy")
```

Coorelation b/w popularity and acousticness on 4% of data

```python
plt.figure(figsize=(10,6))
sns.regplot(data = sample_df, y = "popularity", x="acousticness", color = "Blue").set(title="Correlation between popularity and acousticness")
```

# Duration of songs in different genres

```python
plt.title("Duration of Songs in Different genres")
sns.color_palette("rocket",as_cmap=True)
sns.barplot(y='genre', x='duration', data=df)
plt.xlabel("Duration in Seconds")
plt.ylabel("Genre")
```

# Top 5 Genre by popularity

```python
df_tracks = pd.read_csv('tracks.csv')
```

```python
df_tracks.isnull().sum()
```

```python
df_tracks.fillna('NA')
```

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

# No of songs per year 

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title="Number Of Songs Released Per Year")
```

```python
total_dr=df_tracks.duration_ms
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="Year Vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration_ms
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year Vs Duration")
plt.xticks(rotation=60)
```
